# What is pandas?

-> Pandas is a fast,powerful,flexible and easy to use open source data analysis and manipulation tool,built on top of the python programming language.

# Pandas Series

-> A pandas series is like a coulmn in table.It is a 1D array, holding data of any type.

```python
#importing pandas
import numpy as np
import pandas as pd
```

# Series from lists

```python
#string
country=['India','Pakistan','USA','Nepal','Srilanka']
pd.Series(country)
```

```python
#integers
marks=[92,59,57,43,98]
pd.Series(marks)
```

```python
#custom index
marks=[93,67,87,56,78]
subject=['Math','Cst','DSA','Economics','Digital']

pd.Series(marks,index=subject)
```

```python
#setting a name
pd.Series(marks,index=subject,name='Priya k marks')
```

# Series from dictionaries

```python
marks={
    'math':100,
    'phy':95,
    'chem':96,
    'comp':93
}

pd.Series(marks)
pd.Series(marks,name="Priya Ghosh")
```

# Series Attributes

```python
# size(returns the no. of items present in series )
marks={
    'math':100,
    'phy':95,
    'chem':96,
    'comp':93
}

marks_series=pd.Series(marks,name="Priya")
marks_series.size
```

```python
# dtype(return datatype)
marks_series.dtype
```

```python
# name
marks_series.name
```

```python
# is_unique(it tells that all the items in series are unique or not)
print(marks_series.is_unique)

print(pd.Series([1,1,1,1,3,4,5,2]).is_unique)
```

```python
# index
marks_series.index
```

```python
# type
type(marks_series)
```

```python
# values
a=marks_series.values
print(a)
type(a)
```

# Series Methods

```python
# head(give preview the 1st 5 dataset)
subs=pd.read_csv('C:/Users/pghos/Downloads/subs.csv')
subs_series=subs.iloc[:,0]
print(subs_series.head())
# return 1st two data
print(subs_series.head(2))
```

```python
# value_counts(return the frequency of data)
movie=pd.read_csv('C:/Users/pghos/Downloads/bollywood.csv',index_col='movie')
movie_series=movie.iloc[:,0]
movie_series.value_counts()
```

```python
# sort_values
vk=pd.read_csv('C:/Users/pghos/Downloads/kohli_ipl.csv',index_col='match_no')
vk_series=vk.iloc[:,0]
print(vk_series.sort_values())
# descending order
print(vk_series.sort_values(ascending=False))
print(vk_series.sort_values(ascending=False).head(1))

# inplace function can change original dataset
```

```python
# sort_index
movie=pd.read_csv('C:/Users/pghos/Downloads/bollywood.csv',index_col='movie')
movie_series=movie.iloc[:,0]
movie_series.sort_index()
```

```python
# drop_duplicates
temp=pd.Series([1,2,2,3,3,4,5,5])
temp.drop_duplicates()
```

```python
# duplicated(tells that the duplicate value present in the series or not )
temp.duplicated().sum()
```
