# How to create multiIndex object

1.pd.MultiIndex.from_tuples()

2.pd.MultiIndex.from_product()

```python
# pd.Multiindex_from_tuples()
index_val=[('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
Multiindex=pd.MultiIndex.from_tuples(index_val)

Multiindex
```

```python
Multiindex.levels[0]
```

```python
# pd.MultiIndex.from_product()
pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]])
```

# Creating series with multiple index

```python
series=pd.Series([1,2,3,4,5,6,7,8],index=Multiindex)
series
```

```python
# How to fetch a item from such a series
series['cse']
```

```python
# unstack(convert multiindex series to dataframe)
temp=series.unstack()
temp
```

```python
# stack(convert dataframe to multiindex series)
temp.stack()
```

# Long vs Wide data

->Wide format is where we have a single row for every datapoint with multiple columns to hold the values of various attributes.

->Long format is where we have as many rows as the number of attributes and each row contains the value of a particular attribute for a given data point.

# melt

```python
# melt(covert wide dataframe in to long dataframe)
pd.DataFrame({'cse':[120]}).melt()#simple example branch
```

```python
# melt-> branch with year
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]}).melt()
```

```python
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]}).melt(var_name='branch',value_name='num_student')
```

```python
temp_df=pd.DataFrame(
    {
        'branch':['cse','ece','mech'],
        '2020':[100,150,60],
        '2021':[120,130,70],
        '2022':[150,140,80]
    }
)

temp_df.melt()
```

```python
temp_df.melt(id_vars=['branch'],var_name='year',value_name='num_student')
```

# pivot_table

->The pivot table takes simple column-wise data as input,and groups the entires into two dimensional table that provides a multidimensional summarization of the data. 

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
df.groupby('sex')[['total_bill']].mean()
```

```python
df.groupby(['sex','smoker'])[['total_bill']].mean().unstack()
```
